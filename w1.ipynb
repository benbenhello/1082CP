# 挑戰一
### 折線圖

```python
k = 1000
x = [1,2,3,4]
price =  [100*k,150*k,230*k,80*k]
price_acc = []
for i in range(len(price)):
    if i == 0:
        price_acc.append(price[i])
    else:
        price_acc.append(price[i]+price_acc[i-1])
plot(x,price_acc)
```

### 圓餅圖

```python
plt.pie(price,autopct="%0.1f%%")
```

# 挑戰二

```python
def checkVaildBin(value):
    b = ['0','1']
    binary = True
    for i in value:
        if i not in b:
            binary = False
            break
    return binary
            
while 1:
    switch = input("請選擇服務(輸入 a, b, q)\na:十進位轉二進位\tb:二進位轉十進位\tq:離開\t")
    if switch == 'q':
        print("886")
        break
    elif switch == 'a':
        decimal = input("請輸入十進位數字")
        if decimal.isdigit():
            binary = bin(int(decimal))[2:]
            print("\n{} 轉為二進位後為 {}".format(decimal,binary))
        else:
            print("十進位不懂?\n")
    elif switch == 'b':
        binary = input("請輸入二進位數字")
        if checkVaildBin(binary):
            decimal = int(binary,2)
            print("\n{} 轉為十進位後為 {}".format(binary,decimal))
        else:
            print("二進位不懂?\n")
    else :
        print("中文不通?")
        continue
```
